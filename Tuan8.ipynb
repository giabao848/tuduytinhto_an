```python
class ShoppingCart:
    def __init__(self):
        self.data = []
    def add(self, name, price):
        for item in self.data:
            if item['name'].lower() == name.lower() and item['price'] == price:
                item['count'] += 1
                return
        new_product = {
            'name' : name,
            'price' : price,
            'count' : 1
        }
        self.data.append(new_product)
    def remove(self, name):
        for item in self.data:
            if item['name'].lower() == name.lower():
                self.data.remove(item)
                print(f"Đã xóa {name} khỏi giỏ hàng")
                return
        print("Không tìm thấy sản phẩm")
    def check(self):
        if not self.data:
            print("Giỏ hàng rỗng")
        else:
            print("Giỏ hàng không rỗng")
    def total(self):
        tong = 0
        for item in self.data:
            tong += (int(item['price']) * int(item['count']))
        print(tong)
    def display_ShoppingCart(self):
        for item in self.data:
            print(f"{item['name']} : {item['price']} Số lượng {item['count']}")
    def erase_all(self):
        self.data =[]
        print("Đã xóa toàn bộ giỏ hàng")
print("---- GIỎ HÀNG ----")
print("2.Thêm sản phẩm")
print("3.Xóa sản phẩm")
print("4.Kiểm tra giỏ có rỗng không")
print("5.Tính tổng tiền")
print("6.Hiển thị toàn bộ sản phẩm trong giỏ")
print("7.Xóa toàn bộ giỏ hàng")
cart = ShoppingCart()
while True:
    try:
        choice = input("Bạn muốn làm gì(2-7): ")
        if choice < '2' or choice > '7':
            raise ValueError("Lựa chọn không hợp lệ")
    except ValueError as e:
        print(f"{e}")
        continue
    if choice == '2':
        name = input("Nhập tên sản phẩm: ")
        price = input("Nhập giá sản phầm: ")
        cart.add(name, price)
    elif choice == '3':
        name = input("Nhập tên sản phẩm muốn xóa:")
        cart.remove(name)
    elif choice == '4':
        cart.check()
    elif choice == '5':
        cart.total()
        break
    elif choice =='6':
        cart.display_ShoppingCart()
    elif choice == '7':
        cart.erase_all()
```

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        s = self.w * self.h
        return (f"{s:.2f}")
    def perimeter(self):
        s = (self.w + self.h) * 2
        return (f"{s:.2f}")
    def scale(self, k):
        self.w = k * self.w
        self.h = k * self.h
w, h ,k = map(float, input().split())
result = Rectangle(w, h)
result.scale(k)
print(result.area())
print(result.perimeter())
```

```python
import math
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu khác 0")
        s = math.gcd(num, den)
        self.num = int(num / s)
        self.den = int(den / s)

    def __str__(self):
        return (f"{self.num}/{self.den}")
    def __add__(self, other):
        num_new = self.num * other.den + other.num * self.den
        den_new = self.den * other.den
        return Fraction(num_new, den_new)
    def __sub__(self, other):
        num_new = self.num * other.den - other.num * self.den
        den_new = self.den * other.den
        return Fraction(num_new, den_new)
    def __mul__(self, other):
        num_new = self.num * other.num
        den_new = self.den * other.den
        return Fraction(num_new, den_new)
    def __truediv__(self, other):
        num_new = self.num *  other.den
        den_new = self.den * other.num
        return Fraction(num_new, den_new)
try:
    line = input("Nhập phép tính: ").split()
    a, b = int(line[0]), int(line[1])
    op = line[2]
    c,d = int(line[3]), int(line[4])
    f1 = Fraction(a, b)
    f2 = Fraction(c, d)
except ValueError as e:
    print(f"{e}")
if op == '+':
    print(f1 + f2)
elif op == '-':
    print(f1 - f2)
elif op == '*':
    print(f1 * f2)
elif op == '/':
    print(f1 / f2)
else:
    print("Phép tính không hợp lệ")
```

```python
class BankAccount:
    def __init__(self, owner, balance: float):
        self.owner = owner
        self.balance = float(balance)
    def __str__(self):
        return (f"{self.balance:.2f}")
    def deposit(self, x: float):
        if x <= 0 :
            pass
        else:
            self.balance = self.balance + x
    def withdraw(self, x: float):
        if x > self.balance or x <= 0:
            pass
        else:
            self.balance = self.balance - x
owner, initial_balance = input("Nhập tên và số dư(Cách nhau bởi dấu cách): ").split()
bank = BankAccount(owner, initial_balance)
n = int(input("Nhập số lệnh: "))
while n > 0:
    strs = input("Nhập hành động và giá trị(Cách nhau bới dấu cách): ").split()
    key, value = strs[0], float(strs[1])
    if key.lower() == 'withdraw':
        bank.withdraw(value)
    elif key.lower() == 'deposit':
        bank.deposit(value)
    n -= 1
print(bank)
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.data = {}
    def add_score(self, subject, score: float):
        if 0 <= score <= 10:
            self.data[subject] = score
        else:
            pass
    def avg(self):
        scores = self.data.values()
        if not scores:
            return 0.00
        else:
            return sum(scores) / len(scores)
    def rank(self):
        if self.avg() >= 8:
            return "Excellent"
        elif self.avg() >= 6.5:
            return "Good"
        elif self.avg() >= 5:
            return "Average"
        else:
            return "Poor"
    def __str__(self):
        return f"{self.name} {self.avg()} {self.rank()}"
name = input("Nhập tên: ")
n = int(input("Nhập số môn: "))
student = Student(name)
while n > 0:
    strs = input("Nhập môn và điểm( Cách nhau bởi dấu cách): ").split()
    student.add_score(strs[0], float(strs[1]))
    n -= 1
print(student)
```

```python
import math
class Point:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z
    def __sub__(self, other):
        return Point(other.x - self.x, other.y - self.y, other.z - self.z)
    def vecto(self, other):
        return Point(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
            )
    def abs(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)
    def dot(self, other):
        return (self.x * other.x + self.y * other.y + self.z * other.z)
arr = []
for i in range(4):
    arr.append(Point(*map(float, input().split())))
a, b, c, d = arr
ab = b - a
bc = c - b
bd = d - b
n1 = ab.vecto(bc)
n2 = bc.vecto(bd)
cos_phi = n1.dot(n2) / (n1.abs() * n2.abs())
phi_rad = math.acos(cos_phi)
phi_deg = math.degrees(phi_rad)
print(f"{phi_deg:.2f}")
```

```python
class ComplexNumber:
    def __init__(self, re, im):
        self.re = re
        self.im = im
    def __str__(self):
        return f"{self.re} {self.im}"
    def multi(self, other):
        re_new = self.re * other.re - self.im * other.im
        im_new = self.re * other.im + self.im * other.re
        return ComplexNumber(re_new, im_new)
m = ComplexNumber(*map(int, input().split()))
n = ComplexNumber(*map(int, input().split()))
x = m.multi(n)
print(x)
```

```python
import math
class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3
    def istriangele(self):
        if self.e1 + self.e2 <= self.e3:
            return False
        elif self.e2 + self.e3 <= self.e1:
            return False
        elif self.e1 + self.e3 <= self.e2:
            return False
        return True
    def dientich(self):
        p = (self.e1 + self.e2 + self.e3) / 2
        s = math.sqrt(p * (p - self.e1) * (p - self.e2) * (p - self.e3))
        return f"{s:.2f}"
m = Triangle(*map(int, input().split()))
if not m.istriangele():
    print("invalid")
else:
    print(m.dientich())
```
