W5A3

```python
from math import sqrt
def nguyento(n):
  if n < 2:
    return False
  for i in range(2, int(sqrt(n)) + 1):
    if n % i == 0:
      return False
  return True
a = int(input())
print(nguyento(a))
```

```python
import math
def num(n):
  count = 1
  for i in range(2, int(sqrt(n)) + 1):
    if n % i == 0:
      count += ( i + n/i )
  if count == n:
    return True
  else:
    return False
a = int(input())
print(num(a))
```

```python
def finds(A : list, a: int):
  for i in range(len(A)):
    if A[i] == a:
      return i + 1
  return -1
A = list(map(int, input().split()))
a = int(input())
print(finds(A, a))
```

```python
def cacul(num1: float, operat:str, num2: float):
  if operat == '+':
    kq = num1 + num2
  elif operat == '-':
    kq = num1 - num2
  elif operat == '*':
    kq = num1 * num2
  elif operat == '/':
    if num2 == 0:
      return "Khong the chia cho 0"
    else:
      kq = num1 / num2
  return round(kq,2)
nums = input("Nhập phép tính: ")
import re
part = re.split(r'(\+|\-|\*|/)', nums)
part = [p.strip() for p in part if p.strip()]
i = 0
correct_part = []
while i < len(part):
  item = part[i]
  correct = (item == "-") and (i == 0 or part[i-1] in ["+", "-", "*", "/"])
  if correct:
    correct_part.append(item + part[i+1])
    i +=2
  else:
    correct_part.append(item)
    i += 1
part = correct_part
print(cacul(float(part[0]), part[1], float(part[2])))
```

```python
def hamming(a: int, b: int):
  x = a ^ b
  return bin(x).count('1')
a, b = map(int, input().split())
print(hamming(a, b))
```

```python
def tong(n):
  count = 0
  while n > 0:
    count = count + (n%10)
    n //= 10
  return count
n = int(input())
print(tong(n))
```

```python
def check(a: str, b: str):
  if len(a) != len(b):
    return False
  a_b = {}
  b_a = {}
  for char_a, char_b in zip(a, b):
    if char_a in a_b:
      if a_b[char_a] != char_b:
        return False
    else:
      a_b[char_a] = char_b
  for char_b, char_a in zip(b, a):
    if char_b in b_a:
      if b_a[char_b] != char_a:
        return False
    else:
      b_a[char_b] = char_a
  return True
m, n = map(str, input().split())
print(check(m, n))
```
