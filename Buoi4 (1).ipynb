W4A9

```python
def check(n):
  mask = 0
  while(n > 0):
    d = n % 10
    if mask << d & 1:
      return False
    n //= 10
    mask |= 1 << d
  return True
n = int(input("Nhập số:"))
for i in range(1, int(n**(1/2)) + 1):
  if check(i*i):
    print(i * i, end = " ")
```
